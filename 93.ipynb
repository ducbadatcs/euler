```python
import math

def postfix_eval(l: list[str]) -> float:
    st = []
    for i in l:
        if i.isnumeric():
            st.append(int(i))
        else:
            a, b = st[-1], st[-2]
            st.pop()
            st.pop()
            if i == "+": st.append(a + b)
            elif i == "-": st.append(a - b)
            elif i == "*": st.append(a * b)
            elif i == "/": st.append(a / b)
            
    return int(st[0]) if math.isclose(float(st[0]), int(st[0])) else float(st[0])

l = "1 2 3 4 + + +".split()
postfix_eval(l)
```

```python
def int_expressible(a: int, b: int, c: int, d: int) -> list[int]:
    print(a, b, c, d)
    r: set[int] = set()
    for _i in itertools.product(
        itertools.permutations([a, b, c, d]), 
        itertools.product("+-*/", repeat=3)
    ):
        i, j = _i[0], _i[1]
        si = " ".join(str(_q) for _q in i)
        sj = " ".join(j)
        # print(si, sj)
        l = (si + " " + sj).split()
        b = postfix_eval(l)
        if math.isclose(b, int(b)) and b > 0:
            r.add(int(b))
    # print(r)
    return list(r)
        
int_expressible(1, 5, 6, 7)
```

```python
def mex(l: list[int]) -> int:
    t = 0
    while True:
        t += 1
        if t not in l:
            return t
        
mex(int_expressible(1, 2, 3, 4))
```
